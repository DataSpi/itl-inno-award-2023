# Interact with GPT using a simple prompt

```python
from langchain.prompts import PromptTemplate
from langchain.llms import OpenAI
from langchain.chains import LLMChain

prompt_template = """You are an expert about policies of ITL Corporation, I will ask you a question, and then provide you some chunks of text contain relevant information about the question. 
Try to answer the question based on the text provided in Vietnamese.
Questions: 
{user_question}  

Relevant Information:
{relevant_info}
"""

llm = OpenAI(temperature=0)
llm_chain = LLMChain(
    llm=llm,
    prompt=PromptTemplate.from_template(prompt_template)
)
```

```python
user_question = 'nguyn quc thc l ai?'
relevant_info = """
ni quy lao ng

	01/07/2023

	m ti liu: hr.03.v3.2023

	thm nh

	ph duyt

	nguyn quc thc (simon)
	ph tg  vp nhn s	

	nguyn th kim oanh (kimberly)
	vp khi pht trin bn vng


	trn tun anh (ben)
	tng gim c

	

	son tho & ng thun

	v th ngc mai (jina)
	trng phng tin lng v phc li

	nguyn th phng tho (oliver)
	trng phng nhn s cp cao


	l th hng (minty)
	ph phng php l
	

	nguyn th minh hi (hellen)
	trng phng tun th


	lch s sa i ti liu

	m ti liu

	ngy hiu lc

	ni dung thay i

	trang

	-

	21/02/2019

	cp nht ti liu

	-

	hr.03.v2.2021

	01/04/2021

	cp nht ti liu theo quy nh ca lut lao ng 2019

	ton b

	hr.03.v3.2023

	01/07/2023

	 cp nht ti liu

	ton b

"""

response = llm_chain(
    {
        'user_question': user_question,
        'relevant_info': relevant_info
    }
)
print(response['text'])
```

```python
response = llm_chain(
    {
        'user_question': "ni dung ny c son tho bi ai?",
        'relevant_info': relevant_info
    }
)
print(response['text'])
```
